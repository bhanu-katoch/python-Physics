```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 2000

h = 197.3
m = 940
k=100
#h=1
# m=1
# k=1

r_min=-5
r_max = 5

dd = np.linspace(r_min, r_max, n)
d = dd[2]-dd[1]

f = -(h**2)/(2*m*(d**2))

#m=1
#om=(k/m)**0.5

def V(r,b): return (1/2)*(k*r**2) + (1/3)*b*r**3

bbr=[0,10,30]

aa = []
for j in bbr:
    a = np.zeros([n, n])
    for i in range(n):
        a[i][i] = -2*f + V(dd[i],j)
        if (i == 0):
            a[i][i+1] = 1*f
        elif (i == n-1):
            a[i][i-1] = 1*f
        else:
            a[i][i-1] = 1*f
            a[i][i+1] = 1*f
    aa.append(a)
print(aa[1])
evaluea=[]
eveca=[]
for i in range(len(bbr)):
    evalue, evec = np.linalg.eigh(aa[i])
    evaluea.append(evalue)
    eveca.append(evec)
for i in range(len(bbr)):
    print("\nfor b =",bbr[i])
    for j in range(5):
        print(j+1,"eigen energy :", evaluea[i][j],)
    print()
 
#print(aa[0])
#print(l_(pn)[0])
#print(eveca[1].T[0])
```

```python
dd1 = np.linspace(r_min, r_max, n)

for w in range(len(bbr)):
    for j in range(1):
        U = (-1)**(j+1)*eveca[w].T[j]
        P_d = U**2
      
        plt.subplot(2, 2, 1)
        plt.axhline(y=0, c="black", linestyle="dashed")
        plt.plot(dd1, U, label=(f"U:{j+1}"))
        #plt.xlim([0, 15])
        #plt.ylim([-0.1,0.2])
        #plt.ylim([-0.1,(w+1)**-2])
        plt.legend(loc="best")
        
        plt.subplot(2, 2, 2)
        plt.axhline(y=0, c="black", linestyle="dashed")
        plt.plot(dd1, P_d, label=(f"P:{j+1}"))
        #plt.xlim([0, 15])
        #plt.ylim([-0.001,0.01*(w+2)**-2])
        plt.legend(loc="best")
        
    plt.subplot(2, 2, 3)
    plt.axhline(y=0, c="black", linestyle="dashed")
    plt.plot(dd1, V(dd1,bbr[w]), label="potential")
    #plt.xlim([-5, 5])
    plt.legend(loc="best")
        
    plt.suptitle(("wave function for b: "+str(bbr[w])))
    plt.show()
```
