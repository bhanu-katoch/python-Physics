```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
 #initial conditions
theta=np.pi/4
theta_d=0
t=0
tf=5
dt=0.01
g=9.8
l=0.5
#array for plot
tt1=[]
ttheta=[]
tthetat=[]
tthetatt=[]
```

```python
#fuctions
def ff1(theta,theta_d,t):
    return theta_d
def ff2(theta,theta_d,t):
    return -(g/l)*np.sin(theta)
def theta_dd(theta):
	return -(g/l)*np.sin(theta)
```

```python
#loop(data points)
while t<=tf:
    tt1.append(t)
    tthetat.append(theta_d)
    ttheta.append(theta)
    tthetatt.append(theta_dd(theta))
    theta+=dt*ff1(theta,theta_d,t)
    theta_d+=dt*ff2(theta,theta_d,t)
    t+=dt
a=[1,2]
b=[1,2]
c=[1,2]
d=[1,2]

#data=np.array([[tt1],[ttheta],[tthetat],[tthetatt]])
df = pd.DataFrame({'Time':tt1,'θ':ttheta,'θ`':tthetat,'θ``':tthetatt})
df.iloc[[10]]
df
```

```python
#ploting

plt.plot(tt1,ttheta,label="θ")

plt.plot(tt1,tthetat,label="θ`")

plt.plot(tt1,tthetatt,label="θ``")


plt.xlabel("time")
plt.ylabel(r"$\theta$")
plt.legend(loc="best")
# plt.xlim(-26, 26)
# plt.ylim(-26,26)

plt.title(r"simple pendulum for $\alpha=\frac{\pi}{4}$")

plt.show()
```
